# scale the data
- we need to divide data by 255 so that the pixel range from 0 to 1

```python
x_train=x_train/255
x_test=x_test/255
```

```python
plt.figure(figsize=(15,15))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.imshow(x_train[i])
    plt.colorbar()
    plt.xlabel(class_names[y_train[i]])

```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Flatten,Dropout,BatchNormalization,Activation
from tensorflow.keras.initializers import HeNormal
```

```python
model=Sequential()

# input layer

model.add(Flatten(input_shape=(28,28)))

# Hidden 1

model.add(Dropout(0.2))

model.add(Dense(128,kernel_initializer=HeNormal()))

model.add(BatchNormalization())

model.add(Activation('relu'))

# Hidden 2

model.add(Dropout(0.2))

model.add(Dense(128,kernel_initializer=HeNormal()))

model.add(BatchNormalization())

model.add(Activation('relu'))

# output layer

model.add(Dense(10,activation='softmax',kernel_initializer=HeNormal()))
```

```python
model.summary()
```

## Model Compile

```python
from re import VERBOSE
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
help(model.compile)
```

## model.fit

```python
help(model.fit)
```

```python
history=model.fit(x_train,y_train,batch_size=64,epochs=10,verbose=1,validation_split=0.2)
```

```python
history.history
```

```python
plt.plot(history.history['accuracy'],label='accuracy')
plt.plot(history.history['val_accuracy'],label='val_accuracy')
plt.legend()
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.show()
```

```python
max_probs = np.max(y_pred, axis=1)
pred_indices = np.argmax(y_pred, axis=1)
true_indices = y_test

predicted_classes = [class_names[i] for i in pred_indices]
ground_truth_classes = [class_names[i] for i in true_indices]

df_all_predictions = pd.DataFrame({
    "Max_Probability": max_probs,
    "Predicted_Class": predicted_classes,
    "Predicted_Index": pred_indices,
    "Ground_Truth": ground_truth_classes
})
display(df_all_predictions)
```
